Performs text-based classification for predicting the Cuisine, Course, and Diet categories of a new food item using Multinomial Naive Bayes classifiers trained on TF-IDF vectorized features. The results are then printed and optionally stored in a list.

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import LabelEncoder
import numpy as np


df = df.fillna('')

# Preprocessing
X = df['TranslatedRecipeName'] + ' ' + df['TranslatedIngredients'] + ' ' + df['TranslatedInstructions']
y_cuisine = df['Cuisine']
y_course = df['Course']
y_diet = df['Diet']

label_encoder_cuisine = LabelEncoder().fit(y_cuisine)
label_encoder_course = LabelEncoder().fit(y_course)
label_encoder_diet = LabelEncoder().fit(y_diet)

y_cuisine = label_encoder_cuisine.transform(y_cuisine)
y_course = label_encoder_course.transform(y_course)
y_diet = label_encoder_diet.transform(y_diet)

X_train, X_test, y_cuisine_train, y_cuisine_test, y_course_train, y_course_test, y_diet_train, y_diet_test = train_test_split(
    X, y_cuisine, y_course, y_diet, test_size=0.2, random_state=42
)

vectorizer = TfidfVectorizer()
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

model_cuisine = MultinomialNB().fit(X_train_tfidf, y_cuisine_train)
model_course = MultinomialNB().fit(X_train_tfidf, y_course_train)
model_diet = MultinomialNB().fit(X_train_tfidf, y_diet_train)

new_food_item = "Ragi Dosa"
new_food_item_tfidf = vectorizer.transform([new_food_item])

predicted_cuisine = model_cuisine.predict(new_food_item_tfidf)[0]
predicted_course = model_course.predict(new_food_item_tfidf)[0]
predicted_diet = model_diet.predict(new_food_item_tfidf)[0]
predicted_categories=[]
# Decode numerical labels back to original labels
predicted_cuisine_label = label_encoder_cuisine.inverse_transform([predicted_cuisine])[0]
predicted_course_label = label_encoder_course.inverse_transform([predicted_course])[0]
predicted_diet_label = label_encoder_diet.inverse_transform([predicted_diet])[0]

print("Predicted Cuisine:", predicted_cuisine_label)
print("Predicted Course:", predicted_course_label)
print("Predicted Diet:", predicted_diet_label)
predicted_categories.extend([predicted_cuisine_label,predicted_course_label,predicted_diet_label])
```
